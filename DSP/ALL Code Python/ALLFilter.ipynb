LOW PASS Filter FIR

```python
import scipy.signal as sig
import matplotlib.pyplot as plt
import numpy as np

n=50
fs=8000
fc=1200
b=sig.firwin(n+1,fc,fs=fs,window='hamming',pass_zero='lowpass')
w,h=sig.freqz(b,1,fs=fs)
z,p,k=sig.tf2zpk(b,1)

plt.subplot(3,1,1)
plt.plot(w,np.abs(h))

plt.subplot(3,1,2)
plt.plot(w,np.unwrap(np.angle(h)))

plt.subplot(3,1,3)
plt.scatter(np.real(z),np.imag(z),marker='o',edgecolors='b')
plt.scatter(np.real(p),np.imag(p),marker='x',color='b')
plt.tight_layout()
plt.show()
```

**Low pass Filter IIR**


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

fs=8000
wp=1200/(8000/2)
ws=1500/(8000/2)
rp=1
rs=40
n,w=sig.buttord(wp,ws,rp,rs)
[b,a]=sig.butter(n,w,btype='lowpass')
w,h=sig.freqz(b,a,fs=fs)
z,p,k=sig.tf2zpk(n,a)

plt.subplot(3,1,1)
plt.plot(w,np.abs(h))

plt.subplot(3,1,2)
plt.plot(w,np.unwrap(np.angle(h)))

plt.subplot(3,1,3)
plt.scatter(np.real(z),np.imag(z),marker='o',edgecolors='b')
plt.scatter(np.real(p),np.imag(p),marker='x',color='b')
plt.tight_layout()
plt.show()
```

Given s = np.sin(2 * np.pi * t * 5) + np.sin(2 * np.pi * t * 15) + np.sin(2 * np.pi * t * 30)

```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt

fs=100
t=np.arange(0,1.01,1/fs)
s=np.sin(2*np.pi*t*5)+np.sin(2*np.pi*t*15)+np.sin(2*np.pi*t*30)

wp=np.array([10,20])/(fs/2)
ws=np.array([5,25])/(fs/2)
rp=0.1
rs=40
n,wn=sig.ellipord(wp,ws,rp,rs)
[b,a]=sig.ellip(n,rp,rs,wn,btype='bandpass')
w,h=sig.freqz(b,a,fs=fs)

plt.subplot(3,1,1)
plt.plot(t,s)

plt.subplot(3,1,2)
plt.plot(w,np.abs(h))

plt.subplot(3,1,3)
sf=sig.lfilter(b,a,s)
plt.plot(t,sf)
plt.tight_layout()
plt.grid(True)
plt.show()
```
